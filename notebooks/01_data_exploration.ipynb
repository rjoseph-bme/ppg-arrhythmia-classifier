```python
import wfdb
import matplotlib.pyplot as plt
import os

# This fixes the path issue — tells Python to look one folder up from notebooks
os.chdir('..')

# Load patient record 100 (first recording in the database)
record = wfdb.rdrecord('data/mitdb/100', sampto=3600)

# Load the annotations (labels for each heartbeat)
annotation = wfdb.rdann('data/mitdb/100', 'atr', sampto=3600)

# Plot the signal
plt.figure(figsize=(14, 4))
plt.plot(record.p_signal[:, 0], color='steelblue', linewidth=0.8)
plt.title('MIT-BIH ECG Recording — Patient 100 (First 10 Seconds)')
plt.xlabel('Sample')
plt.ylabel('Amplitude (mV)')
plt.tight_layout()
plt.show()

# Print some basic info
print(f"Signal shape: {record.p_signal.shape}")
print(f"Sample rate: {record.fs} Hz")
print(f"Number of annotated beats: {len(annotation.sample)}")
print(f"Unique beat types: {set(annotation.symbol)}")
```
