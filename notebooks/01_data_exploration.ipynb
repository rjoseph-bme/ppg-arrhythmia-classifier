```python
import wfdb
import matplotlib.pyplot as plt
import os

BASE_DIR = os.path.expanduser('~/Documents/ppg-arrhythmia-classifier')

# Load patient record
record = wfdb.rdrecord(f'{BASE_DIR}/data/mitdb/100', sampto=3600)

# Load annotations
annotation = wfdb.rdann(f'{BASE_DIR}/data/mitdb/100', 'atr', sampto=3600)

# Plot
plt.figure(figsize=(14, 4))
plt.plot(record.p_signal[:, 0], color='steelblue', linewidth=0.8)
plt.title('MIT-BIH ECG Recording — Patient 100 (First 10 Seconds)')
plt.xlabel('Sample')
plt.ylabel('Amplitude (mV)')
plt.tight_layout()
plt.show()

print(f"Signal shape: {record.p_signal.shape}")
print(f"Sample rate: {record.fs} Hz")
print(f"Number of annotated beats: {len(annotation.sample)}")
print(f"Unique beat types: {set(annotation.symbol)}")

plt.figure(figsize=(14, 4))
plt.plot(record.p_signal[:1000, 0], color='steelblue', linewidth=0.8)
plt.title('MIT-BIH ECG — Zoomed In (First ~2.8 Seconds)')
plt.xlabel('Sample')
plt.ylabel('Amplitude (mV)')
plt.tight_layout()
plt.show()
```
